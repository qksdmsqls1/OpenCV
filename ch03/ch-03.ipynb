```python
img = cv2.imread('/content/gdrive/My Drive/Images/lena256.jpg') # 이미지 파일 읽기
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 이미지를 RGB 이미지로 변환
plt.imshow(RGB_img) # 이미지 출력
```

```python
Gray_img = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(Gray_img.shape)
plt.imshow(Gray_img,cmap='gray')
```

```python
cv2_imshow(Gray_img)
```

```python
mlmg =cv2.resize(Gray_img,dsize = (256*4, 256*4))
plt.imshow(mlmg,cmap='gray')
```

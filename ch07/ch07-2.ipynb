```python
#img = cv2.imread('/content/gdrive/My Drive/Images/shrinking.bmp')
img = cv2.imread('/content/gdrive/My Drive/Images/lena256.jpg') # 이미지 파일 읽기

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 이미지를 RGB 이미지로 변환

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
plt.imshow(gimg,cmap='gray')
```

```python
height ,width = gimg.shape[:2]
```

```python
M = np.float32([[1,0,100],[0,1,50]])
outimg = cv2.warpAffine(gimg,M,(width,height))
cv2_imshow(outimg)
```

```python
#M = cv2.getRotationMatrix2D(((width-1)/2.0,(height-1)/2.0),-45,1)
M=cv2.getRotationMatrix2D((0,0),-45,1)
print(M)
print(M.shape)
outimg = cv2.warpAffine(gimg,M,(width,height))
cv2_imshow(outimg)
```

```python
M=cv2.getRotationMatrix2D((0,0),-45,1)
outimg1 = cv2.warpAffine(gimg,M,(width,height))
cv2_imshow(outimg)
```

```python
M = np.float32([[1,0,20],[0,1,20]])
outimg2 = cv2.warpAffine(outimg1,M,(width,height))
cv2_imshow(outimg2)
```

```python
M1 = cv2.getRotationMatrix2D((0,0),-45,1)
M2 = np.float32([[M1[0,0],M1[0,1],M1[0,2]],[M1[1,0],M1[1,1],M1[1,2]],[0,0,1]])
M3 = np.float32([[1,0,20],[0,1,20],[0,0,1]])
M4 = np.dot(M3,M2)
```

```python
M5 = np.float32([[M4[0,0],M4[0,1],M4[0,2]],[M4[1,0],M4[1,1],M4[1,2]]])
outimg3 = cv2.warpAffine(gimg,M5,(width,height))
cv2_imshow(outimg3)
```

```python
M1 = cv2.getRotationMatrix2D((0,0),-45,1)
M2 = np.float32([[M1[0,0],M1[0,1],M1[0,2]],[M1[1,0],M1[1,1],M1[1,2]],[0,0,1]])
M3 = np.float32([[1,0,-(width-1)/2.0],[0,1,-(height-1)/2.0],[0,0,1]])
M4 = np.float32([[1,0,(width-1)/2.0],[0,1,(height-1)/2.0],[0,0,1]])
M5 = np.dot(M2,M3)
M5 = np.dot(M4,M5)
```

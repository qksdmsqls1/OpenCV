```python
#img = cv2.imread('/content/gdrive/My Drive/Images/milkdrop.bmp')
img = cv2.imread('/content/gdrive/My Drive/Images/lena256.jpg') # 이미지 파일 읽기

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 이미지를 RGB 이미지로 변환

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
plt.imshow(gimg,cmap='gray')
```

```python
outimg = cv2.GaussianBlur(gimg,(3,3), 1)
cv2_imshow(outimg)
```

```python
outimg = cv2.GaussianBlur(gimg,(15,15), 1)
cv2_imshow(outimg)
```

```python
outimg = cv2.GaussianBlur(gimg,(15,15), 3)
cv2_imshow(outimg)
```

```python
mean_img = cv2.blur(gimg, (5,5))
cv2_imshow(mean_img)
```

```python
edge_img = cv2.addWeighted(gimg, 1.0 ,mean_img,-1.0,0)
edge_img = edge_img * 5
cv2_imshow(edge_img)
```

```python
out_img = cv2.addWeighted(gimg, 1.0 , edge_img , 1.0 , 0)
cv2_imshow(out_img)
```

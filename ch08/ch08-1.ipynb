```python
#img = cv2.imread('/content/gdrive/My Drive/Images/shrinking.bmp')
img = cv2.imread('/content/gdrive/My Drive/Images/char.bmp') # 이미지 파일 읽기

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 이미지를 RGB 이미지로 변환

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
plt.imshow(gimg,cmap='gray')
```

```python
kernel = np.ones((3,3),np.uint8)
dilateimg = cv2.dilate(gimg , kernel , iterations= 3)
cv2_imshow(dilateimg)
```

```python
edgeimg = dilateimg - gimg
cv2_imshow(edgeimg)
```

```python
#img = cv2.imread('/content/gdrive/My Drive/Images/shrinking.bmp')
img = cv2.imread('/content/gdrive/My Drive/Images/milkdrop.bmp') # 이미지 파일 읽기

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 이미지를 RGB 이미지로 변환

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
plt.imshow(gimg,cmap='gray')
```

```python
kernel = np.ones((3,3),np.uint8)
dilateimg = cv2.dilate(gimg , kernel , iterations= 4)
cv2_imshow(dilateimg)
```

```python
gBin = np.zeros((gimg.shape[0],gimg.shape[1]),dtype = np.ubyte)

threshold = 150

for h in range(gimg.shape[0]):
  for w in range(gimg.shape[1]):
    if(gimg[h,w]> threshold):
       gBin[h,w]= 255

cv2_imshow(gBin)
```

```python
kernel = np.ones((3,3),np.uint8)
erodeimg = cv2.erode(erodeimg, kernel , iterations= 1)
cv2_imshow(erodeimg)
```

```python
kernel = np.ones((3,3),np.uint8)
dilateimg = cv2.dilate(gBin , kernel , iterations= 3)
cv2_imshow(dilateimg)
```

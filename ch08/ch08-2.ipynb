```python
#img = cv2.imread('/content/gdrive/My Drive/Images/shrinking.bmp')
img = cv2.imread('/content/gdrive/My Drive/Images/dong.jpg') # 이미지 파일 읽기

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 이미지를 RGB 이미지로 변환

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
plt.imshow(gimg,cmap='gray')
```

```python
kernel = np.ones((3,1),np.uint8)
dilateimg = cv2.dilate(gimg, kernel , iterations= 2)
cv2_imshow(dilateimg)
```

```python
kernel = np.ones((3,1),np.uint8)
erdoeimg = cv2.erode(gimg, kernel , iterations= 2)
cv2_imshow(erodeimg)
```

```python
#img = cv2.imread('/content/gdrive/My Drive/Images/shrinking.bmp')
img = cv2.imread('/content/gdrive/My Drive/Images/fce(salt_pepper noise).bmp') # 이미지 파일 읽기

RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 이미지를 RGB 이미지로 변환

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
plt.imshow(gimg,cmap='gray')
```

```python
kernel = np.ones((3,3),np.uint8)
dilateimg = cv2.dilate(gimg, kernel , iterations= 1)
cv2_imshow(dilateimg)
```

```python
kernel = np.ones((3,3),np.uint8)
erodeimg = cv2.erode(dilateimg, kernel , iterations= 1)
cv2_imshow(erodeimg)
```
